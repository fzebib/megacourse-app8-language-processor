# Load the Book

```python
with open("miracle_in_the_andes.txt", "r") as file:
    book = file.read()
```

# The most used words (non-articles )

```python
import re
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings
```

```python
d = {}

for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1

d_list = [(value, key) for (key,value) in d.items()]
d_list = sorted(d_list, reverse=True)
d_list[:5]
```

```python
import nltk
filtered_words = []
from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
english_stopwords

for count, word in d_list:
    if word not in english_stopwords:
        filtered_words.append((word, count))

filtered_words
```

# Sentiment Analysis: What is the most positive and most negative chapter?

```python
from nltk.sentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
scores = analyzer.polarity_scores("Hey, look how bad the trees are. I hate them. I really hate them.")
```

```python
if scores["pos"] > scores["neg"]:
    print("If it is a positive text")
else:
    print("It is a negative text")
```

```python
analyzer.polarity_scores(book)
```

# Chapters sentiment analysis

```python
import re 
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
chapters = chapters[1:]

for idx, chapter in enumerate(chapters):
    chapter_scores = {}
    scores = analyzer.polarity_scores(chapter)
    print(scores)
    chapter_scores[idx] = scores

chapter_scores
    
```
